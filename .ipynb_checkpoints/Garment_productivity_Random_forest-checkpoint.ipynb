```python
x=df.drop('actual_productivity',axis=1)
y=df['actual_productivity']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=5)
```

```python
normal_scaler=MinMaxScaler()
x_train_scaled=normal_scaler.fit_transform(x_train)
x_train_df=pd.DataFrame(x_train_scaled,columns=x_train.columns)
x_train_df
```

```python
x_test.reset_index(drop=True,inplace=True)
x_test_scaled=normal_scaler.transform(x_test)
x_test_df=pd.DataFrame(x_test_scaled,columns=x_test.columns)
x_test_df
```

```python
rf_model=RandomForestRegressor(random_state=25)
rf_model.fit(x_train_scaled,y_train)
y_pred=rf_model.predict(x_test_df)
```

```python
mse=mean_squared_error(y_test,y_pred)    # testing
mae=mean_absolute_error(y_test,y_pred)
R2=r2_score(y_test,y_pred)
print(R2)
print(mse)
print(mae)
```

```python
y_pred_train=rf_model.predict(x_train_df)
mse=mean_squared_error(y_train,y_pred_train)    # Training
mae=mean_absolute_error(y_train,y_pred_train)
R2=r2_score(y_train,y_pred_train)
print(R2)
```

```python
rf_model=RandomForestRegressor()
params={'n_estimators':np.arange(1,50),'max_depth':np.arange(2,30),
    'min_samples_split':np.arange(2,30),
    'min_samples_leaf':np.arange(2,15),'criterion':["squared_error", "absolute_error", "poisson"],
    'n_jobs':np.arange(3,15)}
rscv_model=RandomizedSearchCV(rf_model,params,cv=4)
rscv_model.fit(x_train_df,y_train)
rscv_model.best_params_
```

```python
y_pred_rf=rscv_model.predict(x_test_df)
mse=mean_squared_error(y_test,y_pred_rf)    # testing
mae=mean_absolute_error(y_test,y_pred_rf)
R2=r2_score(y_test,y_pred_rf)
print(R2)
print(mse)
print(mae)
```
