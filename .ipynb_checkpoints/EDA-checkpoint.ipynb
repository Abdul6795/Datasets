```python
q1=df['smvpw'].quantile(0.25)
q3=df['smvpw'].quantile(0.75)
iqr=q3-q1
max_limit=q3+3*iqr
max_limit
```

```python
df['smvpw']=np.where(df.smvpw>max_limit,max_limit,df.smvpw)
```

```python
sns.boxplot(df['smvpw'])
```
